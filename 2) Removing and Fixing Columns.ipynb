## 2.2 Changing Columns Casing

```python
import re
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
#df.columns = [x.lower() for x in df.columns]
#df.columns = map.(lambda x: x.lower(), df.columns)
```

```python
df.columns = [re.sub(r'([A-Z])',r'_\1', x).lower() for x in df.columns]
```
