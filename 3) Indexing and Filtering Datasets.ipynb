## 3.4 Filtering data with str.contains

```python
df = pd.read_csv(r'C:\projects\pluralsight\collection\Cleaning Data Python Data Playbook\artwork_data.csv')
df['year'] = pd.to_numeric(df.year, errors='coerce')
```

```python
df.medium
```

```python
type(df.medium)
```

```python
df.medium.str
```

```python
df.medium.str.contains('Graphite')
```

```python
df.loc[df.medium.str.contains('Graphite', na=False), ['artist','medium']]
```

```python
df.loc[df.medium.str.contains('Graphite', na=False, case=False), ['artist','medium']]
```

```python
df.loc[df.medium.str.contains('(?i)Graphite', na=False, case=False, regex=True), ['artist','medium']]
```

```python
df.loc[df.medium.str.contains('(?i)Graphite', na=False, regex=True) | df.medium.str.contains('(?i)Line', na=False, regex=True), ['artist','medium']]
```

```python
df.loc[df.medium.str.contains('Graphite|Line' , na=False, case=False, regex=True), ['artist','medium']]
```
